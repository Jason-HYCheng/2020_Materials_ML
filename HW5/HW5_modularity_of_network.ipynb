```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import PIL.Image as Image
```

```python
# 建立一個圖
G = nx.Graph() # 現在 G 是空的

num_node = 6
# 新增多個節點
for i in range(1,num_node):
    G.add_node(i)
# 新增多個邊
G.add_edges_from([(1,2), (1,3), (2,3), (3,4), (4,5), (4,6), (5,6)])
num_link = G.number_of_edges()
nx.draw(G)
```

```python
deg = []
for i in range(num_node):
    deg.append(G.degree(i+1))
print(deg)
```

```python
def G_modularity(A):
    count = 0
    for i in A:
        SG = G.subgraph(i)
        SGlink = SG.number_of_edges()
        D = 0
        for j in i:
            D += deg[j-1]
        count += SGlink / num_link - (D/2/num_link)**2
    return count
```

```python
Q1 = G_modularity([[1,2,3,4,5,6]])
Q2 = G_modularity([[1],[2],[3],[4],[5],[6]])
Q3 = G_modularity([[1,2,3,6],[4,5]])
Q4 = G_modularity([[1,2,3,4],[5,6]])
Q5 = G_modularity([[1,2,4,6],[3,5]])
Q6 = G_modularity([[1,5],[2,3,4,6]])
Q7 = G_modularity([[1,3],[2,4,5,6]])
Q8 = G_modularity([[1,2,5,6],[3,4]])
```

```python
plt.figure(figsize=(18,18))
plt.subplot(331)
image=Image.open("G1.PNG")
plt.axis('off')
plt.imshow(image)
plt.title("modularity Q ="+str(Q1), fontsize = 16)
plt.subplot(332)
image=Image.open("G2.PNG")
plt.axis('off')
plt.imshow(image)
plt.title("modularity Q ="+str(Q2), fontsize = 16)
plt.subplot(333)
image=Image.open("G3.PNG")
plt.axis('off')
plt.imshow(image)
plt.title("modularity Q ="+str(Q3), fontsize = 16)
plt.subplot(334)
image=Image.open("G4.PNG")
plt.axis('off')
plt.imshow(image)
plt.title("modularity Q ="+str(Q4), fontsize = 16)
plt.subplot(335)
image=Image.open("G5.PNG")
plt.axis('off')
plt.imshow(image)
plt.title("modularity Q ="+str(Q5), fontsize = 16)
plt.subplot(336)
image=Image.open("G6.PNG")
plt.axis('off')
plt.imshow(image)
plt.title("modularity Q ="+str(Q6), fontsize = 16)
plt.subplot(337)
image=Image.open("G7.PNG")
plt.axis('off')
plt.imshow(image)
plt.title("modularity Q ="+str(Q7), fontsize = 16)
plt.subplot(338)
image=Image.open("G8.PNG")
plt.axis('off')
plt.imshow(image)
plt.title("modularity Q ="+str(Q8), fontsize = 16)
```
